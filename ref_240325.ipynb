```python
from qiskit import *
from qiskit.visualization import plot_histogram, plot_bloch_vector
import numpy as np
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
```

```python
qr = QuantumRegister(1, 'qubit')
qc2 = QuantumCircuit(qr)
qc2.initialize((1/np.sqrt(2),1/np.sqrt(2)))
qc2.ry(np.pi/2,[0])
plot_bloch_multivector(Statevector.from_instruction(qc2))

```

```python

qc2.add_register(ClassicalRegister(1, 'classical bit'))
qc2.measure(0,0)
qc2.draw(output='mpl')

```

```python

from qiskit_aer import Aer
# Run the quantum circuit on a statevector simulator backend
backend = Aer.get_backend('statevector_simulator')
job = backend.run(qc2, shots=10000)
count = job.result().get_counts()
count['0'] = 0
plot_histogram(count)
```

```python
qc3=QuantumCircuit(QuantumRegister(1, 'qubit'))
qc3.add_register(ClassicalRegister(2, 'classical bit'))
qc3.initialize((1/np.sqrt(2),1/np.sqrt(2)))
qc3.measure(0,0)

qc3.ry(np.pi/2,0)

qc3.measure(0,1)
qc3.draw(output='mpl')
```

```python

from qiskit_aer import Aer
# Run the quantum circuit on a statevector simulator backend
backend = Aer.get_backend('statevector_simulator')
job = backend.run(qc3, shots=10000)
count = job.result().get_counts()
plot_histogram(count)
```
